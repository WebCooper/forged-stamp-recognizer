## Library Version Check

Checks that the three core libraries are installed and prints their versions:

- **Python** — the base language runtime
- **OpenCV** — used throughout the pipeline for image loading, colour space conversion, morphological operations, and contour detection
- **TensorFlow** — used in notebook 05 for building and training the ResNet50 classifier

Also lists any available GPUs. An empty list `[]` is normal on a CPU-only machine — the full pipeline runs on CPU without any changes needed.

**Expected output:**
```
Python: 3.11.x
OpenCV: 4.x.x
TensorFlow: 2.x.x
Detected GPUs: []
```

````python
import sys
import cv2
import tensorflow as tf
from pathlib import Path

print("Python:", sys.version)
print("OpenCV:", cv2.__version__)
print("TensorFlow:", tf.__version__)
print("Detected GPUs:", tf.config.list_physical_devices("GPU"))
````
